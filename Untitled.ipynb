```python
def smooth(losses, rewards=[]): 
    l = losses/ np.sum(losses)
    r = rewards/ np.sum(rewards)
    re = [] 
    le = [] 
    tot = 0 
    total_le = 0
    for i in range(len(r)): 
        tot += r[i]
        total_le += l[i]
        le.append(total_le)
        re.append(tot)
    og_re = re.copy()
    re = np.array(list(zip(re, np.arange(256))))
    np.array(re).shape
    return le, og_re
```

```python
l1_l, l1_r = smooth(np.abs(losses_l1), losses_l1)
# l2_l, _ = smooth(np.abs(losses_l2), losses_l2)
ssim_l, ssim_r = smooth(losses_ssim, losses_ssim)
```
